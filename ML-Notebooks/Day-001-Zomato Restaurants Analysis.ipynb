```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
zomato = pd.read_csv('../Data/zomato-restaurants-data/zomato.csv',encoding="ISO-8859-1")
country_code = pd.read_excel('../Data/zomato-restaurants-data/Country-Code.xlsx')
data = pd.merge(zomato, country_code, on='Country Code')
#data.head()
```

```python
data.columns
```

```python
f, ax = plt.subplots(1,1, figsize = (15, 4))
#ax = sns.countplot(data['Aggregate rating'])
#There are so many restaurents which are not rated or rated 0. Hence we exclude them as outliers
ax = sns.countplot(data[data['Aggregate rating'] != 0]['Aggregate rating'])
```

```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)

f, ax = plt.subplots(figsize=(16, 16))

# Draw a nested violinplot and split the violins for easier comparison
#sns.violinplot(x="Country", y="Aggregate rating", data=data, split=True, linewidth=2.5,
#               inner="quart")
sns.violinplot(x="Aggregate rating",y="Country",data=data ,palette="Set3", inner="points", bw =.2, cut=2, linewidth=3)
sns.despine(left=True)

f.suptitle('Aggregate rating by Country', fontsize=18, fontweight='bold')
ax.set_xlabel("Country",size = 16,alpha=0.7)
ax.set_ylabel("Aggregate rating",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
f, ax = plt.subplots(figsize=(16, 16))

# Draw a nested violinplot and split the violins for easier comparison
#sns.violinplot(x="Country", y="Aggregate rating", data=data, split=True, linewidth=2.5,
#               inner="quart")
sns.violinplot(x="Aggregate rating",y="Country",data=data ,palette="Set3", inner="points", bw =.2, cut=2, linewidth=3)
sns.despine(left=True)

f.suptitle('Aggregate rating by Country', fontsize=18, fontweight='bold')
ax.set_xlabel("Country",size = 16,alpha=0.7)
ax.set_ylabel("Aggregate rating",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
countries = data['Country'].unique().tolist()
cities = data['City'].unique().tolist()
```

```python
#sns.countplot(data[data['Aggregate rating'] != 0]['Aggregate rating'])
for country in countries:
    f, ax = plt.subplots(figsize=(16, 16))
    sns.violinplot(x="Aggregate rating",y="City",data=data[data['Country']==country] ,palette="Set3", inner="points", bw =.2, cut=2, linewidth=3)
    sns.despine(left=True)
    f.suptitle('Aggregate rating by cities for country '+ country, fontsize=18, fontweight='bold')
    ax.set_xlabel("City",size = 16,alpha=0.7)
    ax.set_ylabel("Aggregate rating",size = 16,alpha=0.7)
    plt.show()
```
